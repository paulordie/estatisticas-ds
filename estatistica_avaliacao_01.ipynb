```python
realMadri = ' Real Madri '
parisSaintGermain = " Paris Saint Germain "

def coefVariation(team):
    return round(100*st.stdev(team)/st.mean(team),2)

varReal = round(st.variance(real_madri),2)
desvioReal = round(st.pstdev(real_madri),2)
meanReal = st.mean(real_madri)
medianReal = st.median(real_madri)
cvReal = coefVariation(real_madri)

varParis = round(st.variance(paris_saint_germain),2)
desvioParis = round(st.pstdev(paris_saint_germain),2)
meanParis = st.mean(paris_saint_germain)
medianParis = st.median(paris_saint_germain)
cvParis = coefVariation(paris_saint_germain)

if desvioParis > desvioReal and meanParis == meanReal:
    print("Desvio padrão do Paris é maior",desvioParis,"ja o real",desvioReal)
else:
    print("O desvio padrão do Paris não é maior - falso")
    
if desvioParis < desvioReal and meanParis != meanReal:
    print("Desvio padrão do Paris é maior",desvioParis,"ja o real",desvioReal)
else:
    print("O desvio padrão do Paris não é maior - falso")
    
if meanParis > meanReal and cvParis == cvReal:
    print("A média do Paris é maior",meanParis,"Ja a média do real", meanReal)
else:
    print("A média e os coeficientes não são iguais")
    
if meanParis < meanReal and cvParis < cvReal:
    print("As médias e o coeficiente de variação do Paris são menores")
else:
    print("As médias e o coeficiente de variação do Paris não são menores")
    
```
